```python
<!DOCTYPE html>
<html lang="en">
<head>
    <meta charset="UTF-8">
    <meta name="viewport" content="width=device-width, initial-scale=1.0">
    <title>Nested Divisions</title>
    <style>
        body {
            font-family: Arial, sans-serif;
            text-align: center;
            margin-top: 50px;
        }
        .parent {
            border: 2px dashed red;
            padding: 20px;
            margin: 10px;
        }
        .sibling {
            border: 2px dashed red;
            padding: 20px;
            margin: 10px;
        }
        .child {
            border: 2px dashed red;
            padding: 20px;
            margin: 10px;
        }
        button {
            padding: 10px 20px;
            font-size: 16px;
            background-color: #4CAF50;
            color: white;
            border: none;
            border-radius: 5px;
            cursor: pointer;
            margin: 10px;
        }
        button:hover {
            background-color: #45a049;
        }
    </style>
</head>
<body>

    <!-- Parent Division -->
    <div class="parent" id="parent">
        Parent Division
        <!-- Sibling Divisions -->
        <div class="sibling" id="sibling1">
            Sibling 1 Division
            <!-- Child Divisions -->
            <div class="child" id="child1">
                Child 1 Division
            </div>
            <div class="child" id="child2">
                Child 2 Division
            </div>
        </div>
        <div class="sibling" id="sibling2">
            Sibling 2 Division
        </div>
    </div>

    <!-- Buttons to change border styles -->
    <button onclick="changeParentBorder()">Parent</button>
    <button onclick="changeSiblingsBorder()">Siblings</button>
    <button onclick="changeChildrenBorder()">Children</button>

    <script>
        // Function to change the border color of the parent division
        function changeParentBorder() {
            var parentDiv = document.getElementById("parent");
            parentDiv.style.borderColor = "blue";  // Change border color to blue
        }

        // Function to change the border style and color of sibling divisions
        function changeSiblingsBorder() {
            var siblingDivs = document.querySelectorAll(".sibling");
            siblingDivs.forEach(function(sibling) {
                sibling.style.borderStyle = "solid";  // Change border style to solid
                sibling.style.borderColor = "green";  // Change border color to green
            });
        }

        // Function to change the border style and color of child divisions
        function changeChildrenBorder() {
            var childDivs = document.querySelectorAll(".child");
            childDivs.forEach(function(child) {
                child.style.borderStyle = "double";  // Change border style to double
                child.style.borderColor = "purple";  // Change border color to purple
            });
        }
    </script>

</body>
</html>
```
